<h2 id="aufbau">Aufbau und Struktur</h2>
Im ersten Beispiel zu Belief Networks wird ein minimales Bayesnetz erstellt. Es gibt einen Knoten *A* und einen Knoten *B*, beide enthalten die Werte `true` und `false`. Es exisiert eine gerichtete Kante von *A* nach *B*.
In der grafischen Darstellung unter dem Code-Beispiel ist die Richtung der Kante am dickeren Ende erkennbar, welche auf den Kindknoten zeigt.

```python
%matplotlib inline
from  primo.networks import BayesianNetwork
from  primo.nodes import DiscreteNode

# Bayesnetz initialisieren
bn = BayesianNetwork()

# Knoten erstellen
A = DiscreteNode("A", ["true","false"])
B = DiscreteNode("B", ["true", "false"])

# Knoten zum Bayesnetz hinzufügen
bn.add_node(A)
bn.add_node(B)

# Eine Kante einfügen
bn.add_edge(A,B)

# Netzwerk in Textform darstellen
print "Netzwerk in Textform:", bn.node_lookup

# Netzwerk grafisch darstellen
bn.draw()
```

<h2 id="cpt">CPTs</h2>
Das nächste Beispiel mit vier Knoten zeigt zusätzlich, wie CPTs zu den Knoten hinzugefügt werden. Diese können entweder direkt als numpy-array gesetzt werden, alternativ werden die Werte für jede Variablenbelegung einzeln übergeben.

```python
%matplotlib inline
from  primo.networks import BayesianNetwork
from  primo.nodes import DiscreteNode
import numpy

# Bayesnetz initialisieren
bn = BayesianNetwork()

# Knoten erstellen
A = DiscreteNode("A", ["true","false"])
B = DiscreteNode("B", ["true", "false"])
C = DiscreteNode("C", ["true", "false"])
D = DiscreteNode("D", ["true", "false"])

# Knoten zum Bayesnetz hinzufügen
bn.add_node(A)
bn.add_node(B)
bn.add_node(C)
bn.add_node(D)

# Kanten einfügen
bn.add_edge(A,B)
bn.add_edge(A,C)
bn.add_edge(B,D)
bn.add_edge(C,D)


# A hängt von keinem weiteren Knoten ab. Ein einfaches Beispiel, wie man den CPT diret als numpy-array setzen kann.
cpt_A = numpy.array([0.001,0.999])
A.set_probability_table(cpt_A,[A])


# Setzen der CPT-Werte für B gegeben A
B.set_probability(0.9,[(A,"true"),(B,"true")])
B.set_probability(0.1,[(A,"true"),(B,"false")])
B.set_probability(0.25,[(A,"false"),(B,"true")])
B.set_probability(0.75,[(A,"false"),(B,"false")])


# Setzen der CPT-Werte für C gegeben A
C.set_probability(0.66,[(A,"true"),(C,"true")])
C.set_probability(0.34,[(A,"true"),(C,"false")])
C.set_probability(0.9,[(A,"false"),(C,"true")])
C.set_probability(0.1,[(A,"false"),(C,"false")])

# Setzen der CPT-Werte für D gegeben B und C
D.set_probability(0.55,[(B,"true"),(C,"true"),(D,"true")])
D.set_probability(0.45,[(B,"true"),(C,"true"),(D,"false")])
D.set_probability(0.23,[(B,"true"),(C,"false"),(D,"true")])
D.set_probability(0.77,[(B,"true"),(C,"false"),(D,"false")])
D.set_probability(0.15,[(B,"false"),(C,"true"),(D,"true")])
D.set_probability(0.85,[(B,"false"),(C,"true"),(D,"false")])
D.set_probability(0.99,[(B,"false"),(C,"false"),(D,"true")])
D.set_probability(0.01,[(B,"false"),(C,"false"),(D,"false")])

# Ausgabe der CPTs (heißt hier cpd...)
print "CPT des Knoten A:", A.cpd, "\n"
print "CPT des Knoten B:", B.cpd, "\n"
print "CPT des Knoten C:", C.cpd, "\n"
print "CPT des Knoten D:", D.cpd, "\n"

# Netzwerk in Textform darstellen
print "Netzwerk in Textform:", bn.node_lookup

# Netzwerk grafisch darstellen
bn.draw()
```

<h2 id="valid">Validität</h2>
Im nächsten Beispiel wird ein Knoten hinzugefügt, welcher ein Netz aus drei Knoten zu einem zyklischen Graphen macht. Der Graph lässt sich zwar erstellen, die Abfrage `is_valid()` gibt jedoch `False` zurück, da das Netz nicht mehr azyklisch (und damit nicht mehr valide) ist.

Weiter werden Möglichkeiten dargestellt, warum Knoten möglicherweise nicht valide sind. Dies ist des Fall, wenn kein CPT gesetzt ist, bzw. dieses Fehler enthält.

```python
%matplotlib inline
from  primo.networks import BayesianNetwork
from  primo.nodes import DiscreteNode
import numpy

# Bayesnetz initialisieren
bn = BayesianNetwork()

# Knoten erstellen
A = DiscreteNode("A", ["true","false"])
B = DiscreteNode("B", ["true", "false"])
C = DiscreteNode("C", ["true", "false"])

# Knoten zum Bayesnetz hinzufügen
bn.add_node(A)
bn.add_node(B)
bn.add_node(C)

# Kanten einfügen
bn.add_edge(A,B)
bn.add_edge(B,C)
bn.add_edge(C,A)

# Für A ist kein CPT gesetzt, daher ist A NICHT valide
print "Ist A valide?", A.is_valid()

# FALSCHES Setzen der CPT-Werte für B gegeben A
# 1. Die Summe der Wahrscheinlichkeiten für einen Fall muss immer 1 sein (hier ist 0.9+0.2=1.1)
B.set_probability(0.9,[(A,"true"),(B,"true")])
B.set_probability(0.2,[(A,"true"),(B,"false")])

# 2. Die Belegungen müssen vollständig definiert sein
B.set_probability(0.25,[(A,"false"),(B,"true")])
# Die Belegungen müssen vollständig definiert sein
#B.set_probability(0.75,[(A,"false"),(B,"false")])


print "Ist B valide?", B.is_valid()


# Setzen der CPT-Werte für C gegeben B
# Hier ist ALLES IN ORDNUNG
C.set_probability(0.66,[(B,"true"),(C,"true")])
C.set_probability(0.34,[(B,"true"),(C,"false")])
C.set_probability(0.9,[(B,"false"),(C,"true")])
C.set_probability(0.1,[(B,"false"),(C,"false")])
print "Ist C valide?", C.is_valid()


# Netzwerk grafisch darstellen
bn.draw()

print "Ist das ein valider Graph?", bn.is_valid()
```

<h2 id="xbif">Aus XMLBIF-Format laden</h2>
XMLBIF ist ein XML-Format für Bayesnetze. Eine detaillierte Erklärung ist hier zu finden: [http://www.cs.cmu.edu/~fgcozman/Research/InterchangeFormat/](http://www.cs.cmu.edu/~fgcozman/Research/InterchangeFormat/)

Das Laden aus einer XMLBIF-Datei verkürzt den Code im Vergleich zu den von Hand erstellten Netzen weiter oben enorm. Die Methode `XMLBIF.read()` gibt direkt ein Netz zurück. Im folgenden Code wird das *Slippery-*Beispiel aus *A. Darwiche - Modeling and Reasoning with Bayesian Networks, S. 57* geladen:


```python
from primo.io import XMLBIF

# Construct a DynmaicBayesianNetwork
bn = XMLBIF.read("xml/slippery.xbif")

# Netzwerk in Textform darstellen
print "Netzwerk in Textform:", bn.node_lookup
```
