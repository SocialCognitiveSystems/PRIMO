<h1>Belief Networks - Approximierte Inferenz</h1>
Mittels MCMC-Verfahren kann in *PRIMO* auch approximierte Inferenz berechnet werden. Hier wird das gleiche Beispiel wie bei der exakten Inferenz verwendet:

````python
%matplotlib inline
from primo.networks import BayesianNetwork
from primo.nodes import DiscreteNode
import numpy

# Netz initialisieren
bn = BayesianNetwork()

# Knoten und deren mögliche Zustände festlegen
burglary = DiscreteNode("Burglary", ["Intruder","Safe"])
alarm = DiscreteNode("Alarm", ["Ringing", "Silent"])
earthquake = DiscreteNode("Earthquake", ["Shaking", "Calm"])
john_calls = DiscreteNode("John calls", ["Calling", "Not Calling"])
baum_calls = DiscreteNode("Baum calls", ["Calling", "Not Calling"])

# Knoten ins Netz hinzufügen
bn.add_node(burglary)
bn.add_node(alarm)
bn.add_node(earthquake)
bn.add_node(john_calls)
bn.add_node(baum_calls)

# Kanten einfügen
bn.add_edge(burglary,alarm)
bn.add_edge(earthquake, alarm)
bn.add_edge(alarm, john_calls)
bn.add_edge(alarm, baum_calls)

# Für Wurzelknoten (ohne Eltern) werden hier im Beispiel die CPTs direkt gesetzt
cpt_burglary = numpy.array([0.001,0.999])
burglary.set_probability_table(cpt_burglary,[burglary])

cpt_earthquake = numpy.array([0.002,0.998])
earthquake.set_probability_table(cpt_earthquake,[earthquake])

# Setzen der CPT-Werte für Knoten mit Eltern
alarm.set_probability(0.95,[(alarm,"Ringing"),(burglary,"Intruder"),(earthquake,"Shaking")])
alarm.set_probability(0.05,[(alarm,"Silent"),(burglary,"Intruder"),(earthquake,"Shaking")])
alarm.set_probability(0.29,[(alarm,"Ringing"),(burglary,"Safe"),(earthquake,"Shaking")])
alarm.set_probability(0.71,[(alarm,"Silent"),(burglary,"Safe"),(earthquake,"Shaking")])
alarm.set_probability(0.94,[(alarm,"Ringing"),(burglary,"Intruder"),(earthquake,"Calm")])
alarm.set_probability(0.06,[(alarm,"Silent"),(burglary,"Intruder"),(earthquake,"Calm")])
alarm.set_probability(0.001,[(alarm,"Ringing"),(burglary,"Safe"),(earthquake,"Calm")])
alarm.set_probability(0.999,[(alarm,"Silent"),(burglary,"Safe"),(earthquake,"Calm")])

baum_calls.set_probability(0.9,[(alarm,"Ringing"),(baum_calls,"Calling")])
baum_calls.set_probability(0.1,[(alarm,"Ringing"),(baum_calls,"Not Calling")])
baum_calls.set_probability(0.05,[(alarm,"Silent"),(baum_calls,"Calling")])
baum_calls.set_probability(0.95,[(alarm,"Silent"),(baum_calls,"Not Calling")])

john_calls.set_probability(0.7,[(alarm,"Ringing"),(john_calls,"Calling")])
john_calls.set_probability(0.3,[(alarm,"Ringing"),(john_calls,"Not Calling")])
john_calls.set_probability(0.01,[(alarm,"Silent"),(john_calls,"Calling")])
john_calls.set_probability(0.99,[(alarm,"Silent"),(john_calls,"Not Calling")])

# Grafische Ausgabe des Netzes
bn.draw()
````

<h2>Funktionen</h2>
<p>
Um MCMC zu initialisieren sind mindestens zwei Parameter notwendig: Das Bayesnetz selbst und die Anzahl der Iterationen.
</p>
<p>Optional können auch das Übergangsmodell und das Limit für den Konvergenztest bei der Berechnung der Markov-Kette gesetzt werden.
Metropolis-Hastings:
```
mcmc = MCMC(bn, 5000, transition_model=MetropolisHastingsTransitionModel()) # DEFAULT
```
Gibbs-Sampling:
```
mcmc = MCMC(bn, 5000, transition_model=GibbsTransitionModel())
```

Limt für den Konvergenztest:
```
mcmc = MCMC(bn, 5000, convergence_test=ConvergenceTestSimpleCounting(500)) # DEFAULT
mcmc = MCMC(bn, 5000, convergence_test=ConvergenceTestSimpleCounting(100))
```

Im Beispiel wird gezeigt, wie mit MCMC a-priori und a-posteriori Wahrscheinlichkeiten, sowie die Evidenzwahrscheinlichkeit (<i>Probability of Evidence</i>) und die MAP-Hypothese berechnet werden können.

````python
from primo.inference.mcmc import MCMC
from primo.inference.mcmc import GibbsTransitionModel
from primo.inference.mcmc import MetropolisHastingsTransitionModel
from primo.evidence import EvidenceEqual as EvEq
from primo.densities import ProbabilityTable
from primo.inference.mcmc import ConvergenceTestSimpleCounting

# MCMC initialisieren: Bayesnetz, Iterationen, transition model
#mcmc = MCMC(bn, 5000, transition_model=GibbsTransitionModel())
#mcmc = MCMC(bn, 5000, transition_model=MetropolisHastingsTransitionModel())
mcmc = MCMC(bn, 5000, transition_model=GibbsTransitionModel(), convergence_test=ConvergenceTestSimpleCounting(500))



print "-------PriorMarginal:-------"
pm = mcmc.calculate_PriorMarginal([alarm],ProbabilityTable)
print "P(Alarm)= " + str(pm)
print "Ground truth=[0.0025  0.9975]\n"

pm = mcmc.calculate_PriorMarginal([burglary],ProbabilityTable)
print "P(Burglary)= " + str(pm)
print "Ground truth=[0.001  0.999]\n"

# Evidenz setzen
evidence = {burglary:EvEq("Intruder")}



print "-------ProbabilityOfEvidence:-------" 
poe = mcmc.calculate_PoE(evidence)
print "p(evidence=Intruder)="+str(poe)
print "Ground truth=0.001\n"

print "-------PosteriorMarginal:-------"
pm = mcmc.calculate_PosteriorMarginal([alarm],evidence,ProbabilityTable)
print "P(alarm|burglary=Intruder)="+str(pm)
print "Ground truth=[0.94  0.06]\n"



print "-------MAP:-------"
hyp = mcmc.calculate_MAP([alarm],evidence,ProbabilityTable)
print "MAP(alarm|burglary=intruder)=" + str(hyp)
print "Ground truth=\"Ringing\"\n"

hyp = mcmc.calculate_MAP([burglary,alarm],{},ProbabilityTable)
print "MAP(burglary,alarm)="+str(hyp)
print "Ground truth=\"Safe\",\"Silent\"\n"
````
